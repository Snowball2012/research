```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from PIL import Image
print(np.__version__)
print(torch.__version__)
```

```python
# load image
# downsample
# linear upsample
# calculate loss (flip)

image = Image.open('data/TCom_Rock_CliffLayered_1.5x1.5_4K_albedo.tif')

#image.show()

imarray = np.array(image)
print(imarray.shape)
plt.imshow(imarray, interpolation='bilinear')


def downsample(image):
    ds_shape = (int(image.shape[0] / 2),int(image.shape[1] / 2), image.shape[2])
    print(ds_shape)
    ds_image = np.empty(ds_shape, dtype=int)
    print(ds_image.shape)
    for i in range(0, int(ds_shape[0])):
        for j in range(0, int(ds_shape[1])):
            base_coord = np.array([i, j]) * 2
            px_sum = np.zeros(ds_shape[2])
            for p in [0,1]:
                for k in [0,1]:
                    px_sum += image[base_coord[0] + p][base_coord[1] + k]
            ds_image[i][j] = px_sum * 0.25
    return ds_image

ds_imarray = downsample(imarray)
plt.imshow(ds_imarray)
```
